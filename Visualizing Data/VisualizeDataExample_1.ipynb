```python
print "Visulaize Data"
train.boxplot()
plt.xticks([1,2,3],["Price","NumberSales","MeanWeight"])
plt.show()
# 价格 # 数量 # 均重
```

```python
price = train.iloc[:,2]  #价格
sns.distplot(price)
plt.xlabel("price")
plt.ylabel("probability")
plt.show()
```

```python
number_sales = train.iloc[:,3] #数量
sns.distplot(number_sales)
plt.xlabel("num_sales")
plt.ylabel("probability")
plt.show()
```

```python
mean_weight = train.iloc[:,4] #均重
sns.distplot(mean_weight)
plt.xlabel("mean_weigth")
plt.ylabel("probability")
plt.show()
```

```python
# visualizing pair relationships
sns.pairplot(train)
plt.show()
```
